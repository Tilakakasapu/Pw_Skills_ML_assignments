```python
# ### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

# - **Probability Mass Function (PMF)**: It is used to describe discrete random variables. The PMF gives the probability that a discrete random variable is exactly equal to some value. For example, when rolling a fair die, the probability that the outcome is 3 is given by the PMF as \( P(X = 3) = \frac{1}{6} \).

# - **Probability Density Function (PDF)**: It is used for continuous random variables. The PDF represents the probability of the variable falling within a particular range, rather than taking an exact value. For example, the height of individuals in a population follows a continuous distribution, and the PDF helps determine the probability that a person's height lies within a specific range (e.g., between 5 and 6 feet).

# ### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why is CDF used?

# - **Cumulative Density Function (CDF)**: The CDF of a random variable is the probability that the variable takes a value less than or equal to a specific value. It is the cumulative sum of the probabilities (or area under the PDF curve for continuous variables). 

#   For example, for a standard normal distribution, the CDF at \( x = 1 \) gives the probability that a randomly selected value from this distribution is less than or equal to 1.

#   **Why is CDF used?** The CDF is useful for determining the probability of a random variable being within a specific range. For example, it helps in calculating the probability that the value of a random variable lies between two points.

# ### Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

# - **Examples of situations where the normal distribution might be used**:
#   1. **Heights of individuals**: The distribution of heights in a population often follows a normal distribution.
#   2. **Measurement errors**: Small measurement errors in scientific experiments often follow a normal distribution.
#   3. **IQ scores**: IQ scores are typically modeled by a normal distribution.

# - **Parameters**: The two parameters of the normal distribution are:
#   - **Mean (\( \mu \))**: The mean determines the center of the distribution. It shifts the graph left or right.
#   - **Standard Deviation (\( \sigma \))**: The standard deviation determines the spread or width of the distribution. A smaller \( \sigma \) means the distribution is narrow, and a larger \( \sigma \) means it is wider.

# ### Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

# - **Importance**: The normal distribution is fundamental in statistics because it is used in many statistical methods, hypothesis testing, and confidence intervals. Many phenomena in nature, science, and business are approximately normally distributed, making it one of the most widely used distributions.

# - **Real-life examples**:
#   1. **Blood pressure**: In a healthy population, blood pressure tends to follow a normal distribution.
#   2. **Test scores**: Many standardized test scores, such as SAT or GRE scores, follow a normal distribution.
#   3. **Weight**: Weights of individuals in a population tend to follow a normal distribution.

# ### Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

# - **Bernoulli Distribution**: It is a discrete probability distribution of a random variable that has exactly two possible outcomes: "success" (usually denoted by 1) or "failure" (denoted by 0). 

#   **Example**: Flipping a fair coin is a Bernoulli trial with outcomes "Heads" (1) or "Tails" (0).

# - **Difference between Bernoulli and Binomial Distribution**:
#   - A **Bernoulli distribution** models a single trial, whereas a **Binomial distribution** models the number of successes in a fixed number of Bernoulli trials.
#   - In a Binomial distribution, there are multiple trials (e.g., 10 coin flips), while in Bernoulli, there is only one trial.

# ### Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

# To calculate this, we can use the **z-score** formula:

# \[
# z = \frac{X - \mu}{\sigma}
# \]

# Where:
# - \( X \) is the value (60),
# - \( \mu \) is the mean (50),
# - \( \sigma \) is the standard deviation (10).

# So:

# \[
# z = \frac{60 - 50}{10} = 1
# \]

# Now, we can use the **z-table** to find the probability corresponding to \( z = 1 \). The area to the left of \( z = 1 \) is approximately 0.8413. 

# Therefore, the probability that \( X > 60 \) is:

# \[
# P(X > 60) = 1 - P(X \leq 60) = 1 - 0.8413 = 0.1587
# \]

# Thus, the probability that a randomly selected observation is greater than 60 is **0.1587** or **15.87%**.

# ### Q7: Explain Uniform Distribution with an example.

# - **Uniform Distribution**: A uniform distribution is a type of probability distribution in which all outcomes are equally likely. The probability of each outcome is the same, meaning the distribution has constant probability over its range.

#   **Example**: Rolling a fair die. Each of the six faces of the die has an equal chance of showing up, so the probability of rolling any number from 1 to 6 is \( \frac{1}{6} \). The distribution is uniform because the outcomes are equally likely.

# ### Q8: What is the z-score? State the importance of the z-score.

# - **Z-score**: A z-score represents the number of standard deviations a data point is from the mean. It is calculated as:

# \[
# z = \frac{X - \mu}{\sigma}
# \]

# Where:
# - \( X \) is the data point,
# - \( \mu \) is the mean,
# - \( \sigma \) is the standard deviation.

# - **Importance**: The z-score is used to standardize data, allowing comparisons between different datasets. It also helps in identifying outliers and calculating probabilities in a standard normal distribution.

# ### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

# - **Central Limit Theorem (CLT)**: The Central Limit Theorem states that the sampling distribution of the sample mean will tend to be normal (or nearly normal) if the sample size is sufficiently large, regardless of the distribution of the population, provided the population has a finite mean and variance.

# - **Significance**: CLT allows statisticians to make inferences about population parameters (like the population mean) using sample data. It is a cornerstone of inferential statistics, enabling hypothesis testing, confidence intervals, and more.

# ### Q10: State the assumptions of the Central Limit Theorem.

# The key assumptions of the Central Limit Theorem are:
# 1. **Independence**: The sampled data points must be independent of each other.
# 2. **Sample Size**: The sample size should be sufficiently large (usually \( n \geq 30 \) is considered large enough).
# 3. **Finite Variance**: The population from which the sample is drawn must have a finite variance.
```
